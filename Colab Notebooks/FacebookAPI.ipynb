```
from robobrowser import RoboBrowser

browser = RoboBrowser()
browser.open('https://facebook.com')

# Get the signup form
signup_form = browser.get_form(name='email')
signup_form         # <RoboForm user[name]=, user[email]=, ...
```

```
email = browser.find(id='email')
passwd = browser.find(id='pass')
```

```
email.text
```

```
# Inspect its values
signup_form['authenticity_token'].value     # 6d03597 ...

# Fill it out
signup_form['user[name]'].value = 'python-robot'
signup_form['user[user_password]'].value = 'secret'

# Submit the form
browser.submit_form(signup_form)
```

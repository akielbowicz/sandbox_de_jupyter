```
from bs4 import BeautifulSoup
from urllib.parse import urlparse 

# sitemap = BeautifulSoup(response.content)
sitemap = BeautifulSoup(response_content)
```

```
import logging 
import gzip

paginas = []

paginas_base = sitemap.find_all('loc')
logging.warn('Hay %s paginas base' %len(paginas_base))


for x in paginas_base:
  logging.warn('Comenzando a parsear: %s' %x.text)
  file = requests.get(x.text)
  g = gzip.decompress(file.content)
  bf = BeautifulSoup(g,'xml')
  links =  [ urlparse(p.text).path for p in bf.find_all('loc') ] 
  logging.warn('Hay %s' %len(links))
  paginas.extend( links )
  logging.warn('Ex %s' %links[:3])
  
  
print( len(paginas)  )
```

# Extracting data

```
import re
from collections import defaultdict

pattern = re.compile(r'\d+')

sections = {}

dates = []
errs = defaultdict(list)

tot_pages = len(list(propiedades()))

for i, p in enumerate(islice(propiedades(),100)):
  try:
    pag = get_page(p)
    soup = BeautifulSoup(pag.content)
    date = soup.find_all(class_='section-date')[0].text
    dates.append(re.search(pattern,date).group())
    sections[p] = soup.find_all('section')
    if i%100==0:
      n = 50*int(i/tot_pages)
      print('\r#'*n,'{}%:{}/{}'.format(n,i,tot_pages),end='')
      print('\r',end='')
  except Exception as e:
    errs[e].append(p)
    print('\r error',e,end='')
```

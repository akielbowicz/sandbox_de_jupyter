````
def gen_search_links():
  yield LINK
  pagina = 2
  while True:
    yield PAGINA.format(pagina=pagina)
    pagina += 1

def get_bs(link):
    response = requests.get(link)
    bs = BeautifulSoup(response.content)
    return bs
    
def properties_links(link):
  bs    = get_bs(link)
  cards = bs.find_all(class_='card')
  links = ( t.attrs['href'] for t in cards )
  yield from links
````

````
all_links = []
for link in gen_search_links():
  links = list(properties_links(link))
  if not links:
    break
  all_links.extend(links)

len(all_links)
````

````
bs = get_bs( LINK )
````

````
def get_card(bs,link):
  nodes = bs.find_all(class_='card') 
  cards = ( t for t in nodes if link == t.attrs['href'] )
  return next(cards, None)

card = get_card(bs, all_links[0])

class Card:
  
  def __init__(self,card):
      self.card        = card
      self.price       = card.find(class_='card__price').text.strip('\r\n $')
      self.location    = card.find(class_='card__location').text.strip('\r\n ')
      self.address     = card.find(class_='card__address').text.strip('\r\n ')
      self.common_data = card.find(class_='card__common-data').text.strip('\r\n ')
      self.title       = card.find(class_='card__title').text.strip('\r\n ')
      self.info        = card.find(class_='card__info').text.strip('\r\n ')
      
  def __str__(self):
    return f'''Price:       {self.price}
             \rLocation:    {self.location}
             \rAddress:     {self.address}
             \rCommon Data: {self.common_data}
             \rTitle:       {self.title}
             \rInfo:        {self.info}
            '''

print(Card(card))
````

````
def get_data(section):
  bullets = section.find_all('li')
  d = {}
  try:
    for b in bullets:
      x,_,y = [ x.string for x in b.contents ]
      d[x] = y
  except ValueError:
      d['extra'] = section.text
    
  return d

def data(details):
  ''' returns a dictionary with common information '''
  sections = details.find_all('section')
  _,_,caracteristicas,basicos,superficie, *otros = sections 

  data = {}

  for section in [caracteristicas,basicos,superficie] + otros:
    data.update( get_data(section) )
  return data

def get_details(bs):
  return bs.find(class_='container details-view')

def get_datos(link):
  pageBs = get_bs( link )
  details = get_details(pageBs)

  direccion   = details.find(class_='property-title-address').text
  descripcion = details.find(class_='description-print').text
  datos = {'link':link, 'direccion': direccion, 'descripcion':descripcion }
  datos.update( data(details) )
  return datos
````

````
all_data = []
errs = []
n = len(all_links)
for i,l in enumerate(all_links):  
  link = BASE_LINK + l
  try:
    all_data.append( get_datos(link) )
  except Exception as e:
    print('error on: ',l)
    errs.append((l,e))
  p =  ( n - i )  / n
  print(f'\r{p:.0%} {"#"*int(100*p)}',end='')
````

````
async def fetch(session,url):
  async with session.get(url) as response:
    return await response.text()

async def main(urls):

  async with aiohttp.ClientSession() as session:
    future_pages = [ fetch(session,url) for url in urls]

#     html = await fetch(session, 'http://python.org' )
    complete, pending = await asyncio.wait(future_pages)
    pages = [t.result() for t in complete ]
    return pages

def get_aio(urls):
  loop = asyncio.new_event_loop()
  content = loop.run_until_complete(main(urls))
  return content
````

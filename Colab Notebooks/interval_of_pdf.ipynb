```
import heapq
from itertools import combinations_with_replacement
```

```
bin = [1,2,3,4,5,4,3,2,1]
total = sum(bin)
bin = [x/total for x in bin]

objective_percentage = 0.8

nodes = filter(lambda n:n[0] < n[1], combinations_with_replacement(range(len(bin)),2))

probability = lambda x : sum(bin[x[0]:x[1]])

expected_probability = lambda x: objective_percentage-0.05 <= x[0] <= objective_percentage+ 0.05
values = filter(expected_probability,
                ( (probability(n),n[1]-n[0],n) for n in nodes) )

values =  sorted( values, key=lambda x: x[0] - x[1] )
```

```
import pandas as pd
import altair as alt
alt.renderers.enable('colab')
```

```
df_X = pd.DataFrame(data=iris.data,columns=iris.feature_names)
df_X.columns = df_X.columns.str.replace('\(cm\)','').str.strip().str.replace(' ','_')
df_X.head()
```

```
df = df_X
s_y = pd.Series(pd.Categorical.from_codes(iris.target, categories=iris.target_names))
df['class'] = s_y
df.head()
```

```
alt.Chart(df).mark_circle().encode(
  alt.X(alt.repeat("column"), type='quantitative'),
  alt.Y(alt.repeat("row"), type='quantitative'),
  color='class:N'  
).properties(
width=150,
height=150
).repeat(
    row=['petal_width','petal_length','sepal_width','sepal_length'],    
    column=['sepal_length','sepal_width','petal_length','petal_width']
)
```

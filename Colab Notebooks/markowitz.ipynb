```
import numpy as np
import matplotlib.pyplot as plt
```

```
expected_return = np.array([0.07, 0.09])
expected_std = np.array([0.05, 0.08])
correlation = 0.3

percentages = np.linspace(0,1,30)

percentages = np.stack((1-percentages,percentages)).T

returns = ( percentages*expected_return ).sum(axis=1)

corr = np.diag(expected_std**2) + np.rot90(correlation*expected_std.prod() * np.eye(2))

stds = np.sqrt( ( percentages @ corr @ percentages.T ).diagonal() )

plt.plot(stds,returns,'o-')
```

```
import altair as alt
import pandas.util.testing as tm
import pandas as pd
alt.renderers.enable('colab')

tm.N,tm.K = 10,3
df = tm.makeDataFrame()
texto = pd.DataFrame({'texto':{'nombre':'Ramiro','link':'www.google.com'},'position':{'nombre':0.2,'link':0}})
```

```
ch = alt.Chart(df).mark_line().encode(
    x=alt.X('A', axis=alt.Axis(format='%', title='porcentaje',labelAngle=45)),
    y='B'
).configure_axis(labelFontSize=12)
ch
```

```
texto = alt.Chart(texto).mark_text(
    align='center',
    baseline='middle'
).encode(
    alt.Y('position:N', axis=alt.Axis(title='', ticks=False,labels=False, grid=False)),
    text='texto:N',
#     href='texto:N'
).configure_text(
    fontSize=20
# href='https://www.google.com'
)
texto
```

```
import altair as alt
import pandas as pd
from numpy import random
```

Notebook de prueba para ver como funciona la selecion interactiva entre varios graficos de inmobiliarias

Quiero ver como teniendo tres graficos:

- Monto vs Superficie y con colores en funcion del numero de ambientrs
- Mapa de las propiedades con tamanio en funcion de la superficie y color con el monto
- Bloques con cada color de la cantidad de ambientes

Al seleccionar el bloque para un numero de ambiente filtro solo ese valor en los otros dos graficos

```
num_propiedades  = 500
df = pd.DataFrame( {'id' :range(num_propiedades),
                    'lat':  50 + 2 * random.random(num_propiedades) - 1,
                    'lon': -55 + 2 * random.random(num_propiedades) - 1,
                    'superficie': 30 * random.random(num_propiedades),
                    'ambientes': random.randint(1,high=7,size=num_propiedades),
                    'monto': 100 * random.random(num_propiedades)
                   } )

# selector = alt.selection_single(empty='all', fields=['id'])

selector = alt.selection_multi(encodings=['color'])


base = alt.Chart(df).add_selection(selector)
```

```
precio = base.mark_circle().encode(
x='superficie:Q',
y='monto:Q',
color=alt.condition(selector, 'ambientes:N', alt.value('lightgray'), legend=None)
)
precio
```

```
mapa = base.mark_circle().encode(
latitude='lat:Q',
longitude='lon:Q',
size='superficie:Q',
color= alt.condition(selector, 'monto:Q',alt.value('lightgray'), legend=None))
mapa
```

```
amb = base.mark_square().encode(
y='ambientes:N',
color=alt.condition(selector, 'ambientes:N', alt.value('lightgray'), legend=None)
)
amb
```
